Pandas series allow for different datatypes as the index.

```python
songs2 = pd.Series([145, 142, 38, 13],
                  name='counts')
print(songs2)
songs2.index
```

The index above is also known as the axis. Because this is a series we only have 1 axis, the index. When using a DataFrame we also have the second index that is usually column names. **Remember that the index is not a part of the values**.

```python
songs3 = pd.Series([145, 142, 38, 13],
                    name='counts',
                   index=['Paul', 'John', 'George', 'Ringo'])
print(songs3)
songs3.index
```

## Categorical Data
If we know our data is limited to a few values we can use categorical data.
- use less memory than strings
- improve performance
- has ordering
- perform operations on categories
- enforce membership on values

```python
s2 = pd.Series(['m', 'l', 'xs', 's', 'xl'], dtype='category')
size_type = pd.api.types.CategoricalDtype(
    categories=['xs', 's', 'm', 'l', 'xl'], ordered=True)
s3 = s2.astype(size_type)
s3
```

```python
"""
1. Using Jupyter, create a series with the temperature values for the last seven days.
Filter out the values below the mean.
"""
temp_series = pd.Series(data=[69.8, 73.4, 75.2, 71.6, 73.94, 71.6, 71.6],
                        index=["M", "Tu", "W", "Th", "F", "Sat", "Sun"],
                        name="7-day forecast")
filter = temp_series > temp_series.mean()
temp_series[filter]      
```
